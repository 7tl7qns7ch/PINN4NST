```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
data_path = "kf_2d_re1000_256_40seed.npy"
data = np.load(data_path)
print(data.shape)
```

```python
num_idx = 0
u = data[num_idx]
print(u.shape)

plt.imshow(u[0], animated=True, vmin=-10, vmax=10, cmap='jet')
plt.colorbar()
```

```python
# create the figure and axes objects
fig, ax = plt.subplots()

# set the initial image
im = ax.imshow(u[0], animated=True, vmin=-10, vmax=10, cmap='jet')

def update(i):
    im.set_array(u[i])
    return im,

# create the animation object
animation_fig = animation.FuncAnimation(fig, update, frames=len(u), interval=10, blit=True, repeat_delay=1,)

# Show the animation
plt.show()

animation_fig.save(f"NS_1000_{num_idx}.gif")
```
