```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
data_path = "data/NS_Re500_s256_T100_test.npy"
data = np.load(data_path)
```

```python
num_idx = 90
u = data[num_idx]
print(u.shape)
```

```python
# create the figure and axes objects
fig, ax = plt.subplots()

# set the initial image
im = ax.imshow(u[0], animated=True, vmin=-20, vmax=10, cmap='jet')

def update(i):
    im.set_array(u[i])
    return im,

# create the animation object
animation_fig = animation.FuncAnimation(fig, update, frames=len(u), interval=20, blit=True, repeat_delay=1,)

# Show the animation
plt.show()

animation_fig.save(f"animation/NS_500_{num_idx}.gif")
```
